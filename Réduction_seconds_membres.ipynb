# On calcule un vecteur "extrema"

La coordonnée i du vecteur est max(coordonnées i des seconds membres) - min(coordonnées i des seconds membres) sur les 1040 seconds membres fournis initialement par RTE 

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import sys
np.set_printoptions(threshold=1000, edgeitems = 100)
```

```python
# On charge les seconds membres
secondmembres = np.vstack(np.array(pd.DataFrame(pd.read_pickle("SecondMember_test.pickle"))[0]))
```

```python
# On calcule vext_max, vext_min et enfin vect_extrema

Nb_obs = 100
n = len(secondmembres[0])

vect_max = []
for ci in secondmembres[0]:
    vect_max.append(ci)

for i in range(Nb_obs):
    for j in range(n) :
        if secondmembres[i][j] > vect_max[j]:
            vect_max[j] = secondmembres[i][j]

vect_min = []
for ci in secondmembres[0]:
    vect_min.append(ci)

for i in range(Nb_obs):
    for j in range(n) :
        if secondmembres[i][j] < vect_min[j]:
            vect_min[j] = secondmembres[i][j]

vect_extrema = []
for i in range(n):
    vect_extrema.append(vect_max[i] - vect_min[i])

vect_extrema = np.array(vect_extrema)
```

```python
# Quelques stats sur les intervalles dans lesquels évoluent les contraintes

compteur0 = 0
for i in range(n):
    if vect_extrema[i] == 0.:
        compteur0 += 1
print(compteur0/n)

# compteurpetit = 0
# petit = 10.
# for i in range(n):
#     if vect_extrema[i] <= petit:
#         compteurpetit += 1
# print(compteurpetit/n)

# compteurgrand = 0
# grand = 1000
# for i in range(n):
#     if vect_extrema[i] >= grand:
#         compteurgrand += 1
# print(compteurgrand/n)
```

```python
# Créer une vecteur extrema_invariant pour donner la numérotation des extrémas qui ne changent pas
extrema_invariant = [];

for i in range(n):
    if vect_extrema[i] == 0.:
        extrema_invariant.append(i)
```

```python
# Même fonction qu'avant pour généner le fichier, on a juste modifié un peu son nom
def genere_seconds_membres_reduced(seconds_membres, noms):
    n = len(seconds_membres[0])
    r = []
    for i in range(len(seconds_membres)):
        assert len(seconds_membres[i]) == n
        r.append((np.array(seconds_membres[i]), noms[i]))
    return r

pickle.dump(genere_seconds_membres_reduced(secondmembres_reduced, noms), open("SecondMember_reduced.pickle","wb"))# remplacer par SecondMembers
```

```python
# Même fonction qu'avant pour lire les seconds membres du fichier "SecondMember_reduced.pickle"
def charge_seconds_membres_reduced():
    import pickle
    l = pickle.load(open("SecondMember_reduced.pickle","rb"))# A changer par l = pickle.load(open("SecondMember.pickle","rb"))
    seconds_membres = []
    noms = []
    for (second_membre, nom) in l:
        seconds_membres.append(second_membre)
        noms.append(nom)
    return (np.array(seconds_membres), noms)
```
